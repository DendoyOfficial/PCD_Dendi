### FUNGSI FLIP

Fungsi flip digunakan untuk membalik citra, baik secara horizontal maupun vertikal, sesuai dengan parameter direction. Prosesnya dimulai dengan mengambil ukuran citra, lalu membuat array kosong dengan dimensi yang sama. Selanjutnya, dua buah perulangan digunakan untuk mengakses setiap elemen pada citra. Setiap elemen kemudian dipindahkan ke posisi baru berdasarkan arah pembalikan yang dipilih: jika direction bernilai horizontal, maka pembalikan dilakukan dengan menukar posisi kolom, sedangkan jika bernilai vertical, pembalikan dilakukan dengan menukar posisi baris. Setelah seluruh elemen diproses, fungsi akan menghasilkan citra yang telah dibalik.


```python
def fungsi_satu(image, mode='X'):
 h, w = image.shape
 hasil = np.zeros((h, w), dtype=image.dtype)
 for i in range(h):
    for j in range(w):
        if mode == 'X':
            hasil[i, j] = image[i, w - 1 - j]
        elif mode == 'Y':
            hasil[i, j] = image[h - 1 - i, j]
 return hasil

rotate1 = fungsi_satu(kicaw1,"Y")
rotate1 = fungsi_satu(rotate1,"X")

plt.imshow(rotate1,cmap="gray")

 
```

Berikut adalah hasil Citra yang telah diflip menggunakan fungsi yang telah dideklrasikan sebelumnya, Citra tersebut diflip ke sumbu Y lalu ke sumbu X.

### FUNGSI ROTASI

Fungsi ini pertama-tama mengukur tinggi dan lebar citra dengan menghitung jumlah baris (tinggi) dan kolom (lebar) citra. Kemudian, fungsi ini membuat array kosong hasil dengan dimensi terbalik, yaitu lebar citra menjadi tinggi dan tinggi citra menjadi lebar. Dalam dua loop yang mengiterasi kolom (j) dan baris (i), elemen-elemen citra dipetakan ke posisi baru sesuai dengan rotasi 90 derajat, di mana elemen yang ada pada posisi (i, j) pada citra asli akan dipindahkan ke posisi (j, tinggi - 1 - i) pada citra yang telah diputar. Setelah semua elemen dipetakan, fungsi mengembalikan array hasil, yaitu citra yang telah diputar 90 derajat.

```python
def cerminBos(image, mode='X'):
 h, w = image.shape
 hasil = np.zeros((h, w), dtype=image.dtype)
 for i in range(h):
    for j in range(w):
        if mode == 'X':
            hasil[i, j] = image[i, w - 1 - j]
        elif mode == 'Y':
            hasil[i, j] = image[h - 1 - i, j]
 return hasil


def rotasi(image, val): 
    h, w = image.shape 
    val = int(val) % 360  
    if val == 0: return image 
    elif val == 90: 
        rot = np.zeros((w, h), dtype=image.dtype) 
        for i in range(h): 
            for j in range(w): rot[j, h - 1 - i] = image[i, j] 
        return rot 
    elif val == 180: 
        rot = np.zeros_like(image, dtype=image.dtype) 
        for i in range(h): 
            for j in range(w): rot[h - 1 - i, w - 1 - j] = image[i, j] 
        return rot 
    elif val == 270:  
        rot = np.zeros((w, h), dtype=image.dtype) 
        for i in range(h): 
            for j in range(w): rot[w - 1 - j, i] = image[i, j] 
        return rot 
    return image 

```

### Fungsi Translasi

Fungsi translasi digunakan untuk menggeser posisi piksel pada citra berdasarkan nilai shift_x dan shift_y. Proses diawali dengan mengubah citra menjadi array NumPy, lalu menentukan ukuran citra berupa tinggi (h) dan lebar (w).

Selanjutnya dibuat array kosong bernama translated dengan ukuran yang sama untuk menyimpan hasil pergeseran. Melalui dua perulangan, setiap piksel diproses dengan menghitung posisi barunya, yaitu menambahkan shift_x pada indeks kolom dan shift_y pada indeks baris.

Jika posisi baru tersebut masih berada dalam batas citra yang valid, maka nilai piksel akan ditempatkan pada posisi tersebut. Setelah seluruh proses selesai, dihasilkan citra baru yang telah bergeser sesuai dengan parameter yang diberikan.

```python
def translasi(img, kolom_x=0, baris_y=0):
    img = np.array(img)
    h, w = img.shape
    hasil = np.zeros ((h,w), dtype = img.dtype)
    for i in range (h):
        for j in range (w):
            newi = i + baris_y
            newj = j + kolom_x
            if 0 <= newi < h and 0 <= newj < w :
                hasil[newi, newj] = img[i, j]
    return hasil

plt.subplot(2,2,1)
neko_1 = translasi (rotate_cuk, 15, 15)
plt.imshow(neko_1, cmap = "gray")

plt.subplot(2,2,2)
neko_2 = translasi (gambar4, -15, 15)
plt.imshow(neko_2, cmap = "gray")

plt.subplot(2,2,3)
neko_3 = translasi (rotate_cuk2, 15, -15)
plt.imshow(neko_3, cmap = "gray")

plt.subplot(2,2,4)
neko_4 = translasi (gambar3, -15, -15)
plt.imshow(neko_4, cmap = "gray")

plt.show()
```

### FUNGSI DILITASI/CROP

Fungsi dilatasi digunakan untuk memperbesar ukuran citra menjadi dua kali lipat dengan cara mereplikasi setiap piksel ke dalam area berukuran 2×2 pada citra baru. Setelah proses pembesaran selesai, citra tersebut dapat dipotong (cropping) dari bagian atas, bawah, kiri, atau kanan sesuai dengan parameter yang diberikan, dengan tetap memastikan bahwa pemotongan tidak melebihi batas dimensi citra.

```python
def dilatasi_crop(image, atas=0, bawah=0, kiri=0, kanan=0):
    h, w = image.shape

    hasil = np.zeros((h*2, w*2), dtype=image.dtype)

    for i in range(h*2):
        for j in range(w*2):
            hasil[i,j] = image[int(i/2), int(j/2)]

    hasil_crop = hasil[atas:h*2-bawah, kiri:w*2-kanan]
    return hasil_crop

cat1 = dilatasi_crop(neko_1, 70, 0, 59, 0)
cat2 = dilatasi_crop(neko_2, 60, 0, 0, 60)
cat3 = dilatasi_crop(neko_3, 0, 60, 60, 0)
cat4 = dilatasi_crop(neko_4, 0, 60, 0, 60)

plt.subplot(2,2,1)
plt.imshow(cat1, cmap='gray')
plt.axis
plt.title("Citra 1")

plt.subplot(2,2,2)
plt.imshow(cat2, cmap='gray')
plt.axis
plt.title("Citra 2")

plt.subplot(2,2,3)
plt.imshow(cat3, cmap='gray')
plt.axis
plt.title("Citra 3")

plt.subplot(2,2,4)
plt.imshow(cat4, cmap='gray')
plt.axis
plt.title("Citra 4")

```

### Penggabungan / Merge 

Fungsi gabung_2x2 digunakan untuk menggabungkan empat citra menjadi satu tampilan dalam bentuk susunan 2×2. Prosesnya diawali dengan menyeragamkan ukuran keempat citra menggunakan fungsi resize_semua, sehingga semua citra memiliki dimensi yang sama berdasarkan ukuran terkecil.

Selanjutnya, fungsi ukuran_max digunakan untuk menentukan ukuran terbesar yang akan dijadikan acuan. Masing-masing citra kemudian dinormalisasi menggunakan fungsi normalisasi, yaitu dengan menyesuaikan ukurannya ke dimensi yang sama dan menambahkan nilai 0 (padding) jika ada bagian yang kurang.

```python

def ukuran_min(gambar):
    baris_min = min(len(citra) for citra in gambar)
    kolom_min = min(len(citra[0]) for citra in gambar)
    return baris_min, kolom_min

def potong(citra, tinggi, lebar):
    return [row[:lebar] for row in citra[:tinggi]]

def resize_semua(gambar):
    tinggi_min, lebar_min = ukuran_min(gambar)
    return [potong(citra, tinggi_min, lebar_min) for citra in gambar]

def ukuran_max(gambar):
    baris_max = max(len(citra) for citra in gambar)
    kolom_max = max(len(row) for citra in gambar for row in citra)
    return baris_max, kolom_max

def normalisasi(citra, baris_tujuan, kolom_tujuan):
    hasil = []
    for i in range(baris_tujuan):
        baris = []
        for j in range(kolom_tujuan):
            if i < len(citra) and j < len(citra[i]):
                baris.append(citra[i][j])
            else:
                baris.append(0)
        hasil.append(baris)
    return hasil

def gabung_2x2(citra1, citra2, citra3, citra4):
    citra1, citra2, citra3, citra4 = resize_semua([citra1, citra2, citra3, citra4])

    b, k = ukuran_max([citra1, citra2, citra3, citra4])
    i1 = normalisasi(citra1, b, k)
    i2 = normalisasi(citra2, b, k)
    i3 = normalisasi(citra3, b, k)
    i4 = normalisasi(citra4, b, k)

    gabung = []
    for i in range(b * 2):
        baris = []
        for j in range(k * 2):
            if i < b and j < k:
                baris.append(i1[i][j]) 
            elif i < b and j >= k:
                baris.append(i2[i][j - k]) 
            elif i >= b and j < k:
                baris.append(i3[i - b][j])
            else:
                baris.append(i4[i - b][j - k]) 
        gabung.append(baris)

    return gabung


cat_real = gabung_2x2(cat1, cat2, cat3, cat4)

plt.imshow(cat_real, cmap="gray")
plt.show()
```

### Fungsi hitung_histogram

Fungsi ini digunakan untuk menghitung distribusi intensitas piksel pada citra. Proses diawali dengan mengubah gambar menjadi array NumPy. Jika citra berwarna (memiliki 3 kanal), maka dikonversi ke grayscale dengan mengambil rata-rata tiap kanal. Selanjutnya, jika nilai piksel masih berada pada rentang 0–1, nilainya diubah ke skala 0–255.

Setelah itu, seluruh piksel diratakan menjadi satu dimensi, lalu dihitung frekuensi kemunculan setiap nilai intensitas (0–255) menggunakan array histogram dengan 256 elemen. Hasil akhirnya berupa distribusi intensitas piksel.

### Fungsi normalisasi_citra
Fungsi ini digunakan untuk menyesuaikan rentang nilai piksel agar berada pada skala 0–255. Pertama, citra diubah menjadi array bertipe float, kemudian dicari nilai minimum (f_min) dan maksimum (f_max).

Jika semua piksel memiliki nilai yang sama, citra cukup dikurangi dengan nilai minimum. Jika tidak, dilakukan normalisasi menggunakan rumus skala linier agar nilai piksel tersebar dari 0 hingga 255. Hasil akhir dikembalikan dalam bentuk integer

```python
def hitung_histogram(gambar):
    img_arr = np.array(gambar)
    
    if len(img_arr.shape) == 3:
        img_arr = np.mean(img_arr, axis=2)
        
    if img_arr.max() <= 1.1:
        img_arr = img_arr * 255
        
    piksel_flat = img_arr.flatten().astype(int)
    
    histogram = [0] * 256
    for p in piksel_flat:
        if 0 <= p <= 255:
            histogram[p] += 1
    return histogram

def normalisasi_citra(gambar):
    img_array = np.array(gambar).astype(float)
    
    f_min = np.min(img_array)
    f_max = np.max(img_array)
    
    if f_max - f_min == 0:
        img_norm = img_array - f_min
    else:
        img_norm = ((img_array - f_min) / (f_max - f_min)) * 255
    
    return img_norm.astype(np.uint8)

def tampilkan_perbandingan_histogram(gambar_asli, gambar_norm):
    hist_asli = hitung_histogram(gambar_asli)
    hist_norm = hitung_histogram(gambar_norm)

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

    ax1.bar(range(256), hist_asli, color='blue', width=1)
    ax1.set_title("Histogram Asli", fontsize=14)
    ax1.set_xlabel("Nilai Intensitas (0-255)")
    ax1.set_ylabel("Jumlah Piksel")
    ax1.grid(True, linestyle='--', alpha=0.5)

    ax2.bar(range(256), hist_norm, color='green', width=1)
    ax2.set_title("Histogram Setelah Normalisasi", fontsize=14)
    ax2.set_xlabel("Nilai Intensitas (0-255)")
    ax2.set_ylabel("Jumlah Piksel")
    ax2.grid(True, linestyle='--', alpha=0.5)

    plt.tight_layout()
    plt.show()

kicaww = normalisasi_citra(cat_real)

tampilkan_perbandingan_histogram(cat_real, kicaww)
```

### FUNGSI KONVOLUSI

Fungsi konvolusi digunakan untuk memproses citra dengan menerapkan filter tertentu melalui operasi konvolusi menggunakan kernel. Pada awalnya, citra diubah menjadi array NumPy bertipe float agar mudah dihitung. Jika citra masih berwarna, maka akan dikonversi menjadi grayscale dengan menggabungkan nilai kanal RGB menggunakan bobot tertentu. Setelah itu, ukuran citra dan kernel ditentukan, lalu dilakukan penambahan padding (nilai nol di sekeliling citra) agar proses konvolusi tetap bisa dilakukan hingga bagian tepi.

Selanjutnya, setiap piksel pada citra diproses menggunakan perulangan dengan cara mengambil area sesuai ukuran kernel, kemudian dikalikan dan dijumlahkan untuk menghasilkan nilai baru. Hasil konvolusi ini kemudian dinormalisasi ke rentang 0–255 agar dapat ditampilkan dengan baik sebagai citra. Jika terdapat perbedaan nilai minimum dan maksimum, dilakukan normalisasi linier; jika tidak, cukup dilakukan penyesuaian sederhana. Hasil akhirnya adalah citra baru yang telah mengalami proses filtering sesuai kernel yang digunakan.

```python
def konvolusi(gambar, kernel):
    img_arr = np.array(gambar).astype(float)
    if len(img_arr.shape) == 3:
        img_arr = 0.299 * img_arr[:,:,0] + 0.587 * img_arr[:,:,1] + 0.114 * img_arr[:,:,2]
    
    tinggi, lebar = img_arr.shape
    k_tinggi, k_lebar = kernel.shape
    
 
    pad_h = k_tinggi // 2
    pad_w = k_lebar // 2
    img_padded = np.pad(img_arr, ((pad_h, pad_h), (pad_w, pad_w)), mode='constant', constant_values=0)
    
    output = np.zeros_like(img_arr)
    
    for i in range(tinggi):
        for j in range(lebar):
            area = img_padded[i:i+k_tinggi, j:j+k_lebar]
            nilai_baru = np.sum(area * kernel)
            output[i, j] = nilai_baru
            
    f_min = output.min()
    f_max = output.max()
    
    if f_max - f_min != 0:
        output_normalized = ((output - f_min) / (f_max - f_min)) * 255
    else:
        output_normalized = output - f_min
        
    return output_normalized.astype(np.uint8)


kernel_kicau = np.array([
    [-1/10, -1/10, -1/10, -1/10],
    [-1/10,  4/10,  4/10, -1/10],
    [-1/10,  4/10,  4/10, -1/10],
    [-1/10, -1/10, -1/10, -1/10]
])

Kicau_Mania_Filtered = konvolusi(kicaww, kernel_kicau)

tampilkan_perbandingan_histogram(kicaww, Kicau_Mania_Filtered)

plt.imshow(Kicau_Mania_Filtered, cmap="gray")
plt.title("Hasil Zero Padding")
plt.show()
```
