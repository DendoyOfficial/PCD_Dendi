```python
def fungsi_satu(image, mode='X'):
 h, w = image.shape
 hasil = np.zeros((h, w), dtype=image.dtype)
 for i in range(h):
    for j in range(w):
        if mode == 'X':
            hasil[i, j] = image[i, w - 1 - j]
        elif mode == 'Y':
            hasil[i, j] = image[h - 1 - i, j]
 return hasil

rotate1 = fungsi_satu(kicaw1,"Y")
rotate1 = fungsi_satu(rotate1,"X")

plt.imshow(rotate1,cmap="gray")

 
```

```python
def cerminBos(image, mode='X'):
 h, w = image.shape
 hasil = np.zeros((h, w), dtype=image.dtype)
 for i in range(h):
    for j in range(w):
        if mode == 'X':
            hasil[i, j] = image[i, w - 1 - j]
        elif mode == 'Y':
            hasil[i, j] = image[h - 1 - i, j]
 return hasil

#gabung3_2 dan gabung4_2

gambar3 = cerminBos(gabung_3_2,'X')
gambar4 = cerminBos(gabung_4_2,'X')

plt.subplot(2, 2, 1)
plt.imshow(gambar3, cmap="gray")

plt.subplot(2, 2, 2)
plt.imshow(gambar4, cmap="gray")
```

```python
def fungsi_empat_R(image, val):
 h, w = image.shape
 val = int(val) % 360
 if val == 0: return image
 elif val == 90:
    rot = np.zeros((w, h), dtype=image.dtype)
    for i in range(h):
        for j in range(w): rot[j, h - 1 - i] = image[i, j]
        return rot
 elif val == 180:
    rot = np.zeros_like(image, dtype=image.dtype)
    for i in range(h):
        for j in range(w): rot[h - 1 - i, w - 1 - j] = image[i, j]
        return rot
 elif val == 270:
    rot = np.zeros((w, h), dtype=image.dtype)
    for i in range(h):
        for j in range(w): rot[w - 1 - j, i] = image[i, j]
        return rot
    return image


#gabung1_2 dan gabung2_2

rotate_cuk = fungsi_empat_R(gabung_1_2,30)
rotate_cuk2 = fungsi_empat_R(gabung_2_2,30)

plt.subplot(2, 2, 1)
plt.imshow(rotate_cuk, cmap="gray")

plt.subplot(2, 2, 2)
plt.imshow(rotate_cuk2, cmap="gray")
```
